## 3. Load the SDXL pipeline

**Model:** [Juggernaut XL v9](https://huggingface.co/RunDiffusion/Juggernaut-XL-v9) â€” best free photoreal SDXL checkpoint, no auth needed.

**Why not Pony Realism?** Pony's HF mirror is gated; Juggernaut XL v9 is fully public and produces better photoreal portraits anyway. We'll swap to a custom LoRA-fine-tuned version once Civitai training is done.

**VRAM:** ~10 GB on T4 in fp16 â€” fits comfortably.

```python
import torch
from diffusers import StableDiffusionXLPipeline, EulerAncestralDiscreteScheduler

MODEL_ID = 'RunDiffusion/Juggernaut-XL-v9'

pipe = StableDiffusionXLPipeline.from_pretrained(
    MODEL_ID,
    torch_dtype=torch.float16,
    variant='fp16',
    use_safetensors=True,
).to('cuda')

pipe.scheduler = EulerAncestralDiscreteScheduler.from_config(pipe.scheduler.config)
try:
    pipe.enable_xformers_memory_efficient_attention()
    print('xformers enabled.')
except Exception:
    pipe.enable_attention_slicing(1)
    print('xformers unavailable, using attention slicing.')
pipe.enable_vae_slicing()

print('Pipeline loaded. Ready to generate.')
```

## 5. Persona prompt configuration

These mirror `persona/style.md` â€” keep them in sync if you tweak.

```python
MASTER_POSITIVE = (
    'photorealistic, masterpiece, best quality, highly detailed face, sharp focus on eyes, '
    '22 year old woman, fair skin, long dark wavy hair, gentle melancholic expression, '
    '{subject}, '
    'gothic dark academia fashion, victorian-inspired clothing, lace high collar, silver locket, '
    'candlelit interior, deep chiaroscuro lighting, warm amber highlights, cool teal shadows, '
    'shallow depth of field, bokeh, 35mm film grain, desaturated colour grade'
)

MASTER_NEGATIVE = (
    '3d, plastic skin, doll, cartoon, anime, illustration, painting, '
    'bright colours, neon, oversaturated, overexposed, harsh shadows, '
    'modern clothing, streetwear, sportswear, athleisure, jeans, t-shirt, '
    'smiling teeth, glasses, sunglasses, '
    'multiple light sources, fluorescent lighting, direct sunlight on face, beach, summer, '
    'text, watermark, signature, logo, '
    'blurry, low quality, lowres, jpeg artifacts, '
    'deformed hands, extra fingers, missing fingers, deformed eyes, asymmetric eyes, '
    'two heads, extra limbs, asymmetric face, '
    'old woman, child, man'
)

# 8 scene types (matches persona/style.md). Generate 15 variations per scene = 120 images.
SUBJECTS = [
    ('library', 'sitting at antique wooden desk, leather-bound book open, brass three-arm candleholder, stained glass window behind, reading pose, fingers tracing text on parchment'),
    ('forest', 'walking through misty autumn woodland, fallen leaves, hooded velvet cloak, single candle lantern, dusk light through bare branches, contemplative gaze, mid-shot'),
    ('tarot', 'holding single tarot card up to candlelight, smith-waite deck, silver rings, dramatic shadow on face, ritual table with crystals and dried flowers'),
    ('outfit_grwm', 'full-body shot, gothic victorian dress, lace high collar, corset detail, dark velvet fabric, standing by tall arched window, side-light, mirror reflection partial'),
    ('study_close', 'close-up of hands, journal page, fountain pen mid-stroke, ink stains, dried pressed flowers, wax-sealed envelope, warm desk lamp, tarot card as bookmark'),
    ('books_flatlay', 'top-down 45-degree flatlay of stack of leather-bound books on velvet cloth, single lit candle, raven feather quill, autumn leaves scattered, no person'),
    ('church', 'seated in empty gothic church pew, candle in hand, stained glass casting coloured light on face, contemplative, hands folded, single shaft of warm light from window'),
    ('bedroom', 'seated on canopy bed with velvet curtains, surrounded by books and candles, reading in nightgown, dim warm lamp, black cat curled on coverlet')
]

VARIATIONS_PER_SCENE = 15  # total = 8 Ã— 15 = 120 images
WIDTH, HEIGHT = 832, 1216   # SDXL 2:3 portrait
STEPS = 28
CFG = 6.5

print(f'Will generate {len(SUBJECTS) * VARIATIONS_PER_SCENE} images at {WIDTH}x{HEIGHT}.')
```

## 6. Generate (~25â€“35 min on T4)

Each image saves as `{scene}_{seed:06d}.png` in your Drive output folder. You'll review them in Drive, pick the 30 with the most consistent character (same face, similar hair, similar build), and submit *those 30* to Civitai's free LoRA trainer.

```python
import random
from PIL import Image

random.seed(42)
total = len(SUBJECTS) * VARIATIONS_PER_SCENE
done = 0
generator = torch.Generator('cuda')  # create once, re-seed each image

for scene_name, subject in SUBJECTS:
    print(f'\n=== {scene_name.upper()} ===')
    for i in range(VARIATIONS_PER_SCENE):
        seed = random.randint(0, 2_000_000_000)
        generator.manual_seed(seed)
        prompt = MASTER_POSITIVE.format(subject=subject)
        
        image = pipe(
            prompt=prompt,
            negative_prompt=MASTER_NEGATIVE,
            width=WIDTH,
            height=HEIGHT,
            num_inference_steps=STEPS,
            guidance_scale=CFG,
            generator=generator,
        ).images[0]
        
        out_path = f'{OUT_DIR}/{scene_name}_{seed:010d}.png'
        image.save(out_path)
        done += 1
        print(f'  [{done:>3}/{total}] {scene_name} seed={seed}')
    torch.cuda.empty_cache()

print(f'\nDone. {done} images saved to {OUT_DIR}')
```

## 8. Selection criteria (you do this offline)

Open the zip on your PC. Build a `picks/` folder. Move the **30 images** that satisfy ALL of:

1. **Same face.** Eye shape, nose, jaw should look like the same person across all 30 picks. Discard ones where the face drifts.
2. **Same hair.** Length, colour, parting. Wavy and dark-brown is the spec.
3. **Same body type.** Slim, average height.
4. **Clean anatomy.** No deformed hands, weird eyes, extra fingers.
5. **Aesthetic on-brand.** Candlelit, gothic, melancholy â€” not modern or bright.
6. **Variety of poses + environments.** At least 4 scene types represented; mix of full-body, mid-shot, close-up, and flatlay.

Upload the 30 picks to **Civitai's free LoRA trainer** ([civitai.com/training](https://civitai.com/training)) â€” it's a single drag-drop interface. ~30 min in queue, ~1 hr training. Output is a `.safetensors` file. Save to Drive, swap into Cell 4 of this notebook, and every future generation uses the locked character.

**Cost: free.** (Civitai gives every account a daily training credit budget; one LoRA fits in the free quota.)
