## 3. Load AnimateDiff pipeline (~2 min first run)

Base model: `epiCRealism` â€” a photorealistic SD1.5 checkpoint. Gives cleaner skin and more realistic faces than the default SD1.5, while still carrying AnimateDiff's smooth motion.

Motion adapter: `animatediff-motion-adapter-v1-5-3` â€” the current recommended stable version.

```python
import torch
from diffusers import AnimateDiffPipeline, DDIMScheduler, MotionAdapter
from diffusers.utils import export_to_video

DEVICE = 'cuda'

print('Loading motion adapter...')
adapter = MotionAdapter.from_pretrained(
    'guoyww/animatediff-motion-adapter-v1-5-3',
    torch_dtype=torch.float16
)

print('Loading base model (epiCRealism)...')
pipe = AnimateDiffPipeline.from_pretrained(
    'emilianJR/epiCRealism',
    motion_adapter=adapter,
    torch_dtype=torch.float16
).to(DEVICE)

pipe.scheduler = DDIMScheduler.from_config(
    pipe.scheduler.config,
    beta_schedule='linear',
    clip_sample=False,
    timestep_spacing='linspace',
    steps_offset=1
)
pipe.enable_vae_slicing()
pipe.enable_attention_slicing(1)

print('Pipeline ready.')
```

## 5. Cordelia prompt library for AnimateDiff

**Note:** AnimateDiff uses SD1.5 â€” prompts are shorter and simpler than SDXL. No `score_9` tags. Focus on subject + mood + lighting.

```python
# Quality prefix used for ALL clips (character and b-roll alike)
QUALITY_BASE = (
    'photorealistic, masterpiece, best quality, cinematic, '
    'gothic dark academia, candlelit, '
    'chiaroscuro lighting, shallow depth of field, film grain, muted colour grade'
)

# Character addition — prepended only for clips featuring Cordelia
CHARACTER = (
    '22 year old woman, fair skin, long dark wavy hair, melancholic expression, '
    'victorian clothing, lace collar'
)

BASE = f'{QUALITY_BASE}, {CHARACTER}'  # full character prompt

NEGATIVE = (
    'cartoon, anime, 3d, plastic, bright colours, neon, modern clothing, '
    'smiling, text, watermark, blurry, low quality, deformed, extra limbs'
)

# Scene clips — each generates one ~2 second atmospheric clip
SCENES = [
    ('library',      f'{BASE}, sitting at antique desk, leather book open, candle flame flickering, stained glass window'),
    ('forest',       f'{BASE}, walking through misty autumn woodland, hooded cloak, bare branches, lantern light, dusk'),
    ('tarot',        f'{BASE}, holding tarot card to candlelight, silver rings, dramatic shadow, crystals on ritual table'),
    ('study',        f'{QUALITY_BASE}, close-up of hands writing in journal, fountain pen, ink, pressed flowers, warm lamp, dark academia, no face'),
    ('books',        f'{QUALITY_BASE}, stack of leather books on velvet, single candle, raven feather, autumn leaves, top-down shot, no person'),
    ('church',       f'{BASE}, seated in gothic church pew, stained glass light on face, candle in hand, contemplative'),
    ('bedroom',      f'{BASE}, seated on canopy bed, books and candles, reading by lamplight, black cat on coverlet'),
    ('candle_broll', f'{QUALITY_BASE}, single brass three-arm candleholder, three candles burning, melting wax, dark background, close-up, warm amber light'),
]

print(f'{len(SCENES)} scenes loaded.')
```

## 6. Generate all clips (~3â€“5 min per clip)

Each clip saves as `{scene_name}_{seed}.mp4` in your output folder.

```python
import random

random.seed(42)
total = len(SCENES)
generator = torch.Generator(DEVICE)  # create once, re-seed each clip

for i, (scene_name, prompt) in enumerate(SCENES):
    seed = random.randint(0, 2_000_000_000)
    generator.manual_seed(seed)
    
    print(f'[{i+1}/{total}] Generating: {scene_name} (seed={seed})')
    
    output = pipe(
        prompt=prompt,
        negative_prompt=NEGATIVE,
        num_frames=16,
        width=512,
        height=768,     # portrait 2:3
        guidance_scale=7.5,
        num_inference_steps=25,
        generator=generator,
    )
    
    frames = output.frames[0]
    del output  # free latent tensors before export
    out_path = f'{OUT_DIR}/{scene_name}_{seed:010d}.mp4'
    export_to_video(frames, out_path, fps=8)
    del frames
    torch.cuda.empty_cache()
    print(f'  Saved: {out_path}')

print(f'\nDone. {total} clips in {OUT_DIR}')
```

```python
import torch

# --- EDIT THESE ---
PROMPT   = f'{BASE}, sitting at antique desk, leather book open, candle flame'
SEED     = 12345
OUT_NAME = 'test_clip'
# ------------------

generator = torch.Generator(DEVICE).manual_seed(SEED)
frames = pipe(
    prompt=PROMPT,
    negative_prompt=NEGATIVE,
    num_frames=16,
    width=512,
    height=768,
    guidance_scale=7.5,
    num_inference_steps=25,
    generator=generator,
).frames[0]

out_path = f'{OUT_DIR}/{OUT_NAME}_{SEED}.mp4'
export_to_video(frames, out_path, fps=8)
print(f'Saved: {out_path}')
```

## Notes

- **Resolution:** 512Ã—768 is the safe SD1.5 size. Going larger (768Ã—1152) is possible but uses more VRAM and may OOM on free T4. If you get OOM, reduce back to 512Ã—768.
- **Frame count:** 16 frames at 8fps = 2 seconds. Increase to 24 frames for 3 seconds, but generation time scales linearly.
- **Chaining clips:** import 3 clips into CapCut as separate tracks â†’ trim tight â†’ crossfade 0.3s between them â†’ you have a 6-second atmospheric intro.
- **B-roll clips** (books, candle): no face = no consistency issues. Generate as many of these as you want for filler content between character posts.
- **When to use animated vs realistic:** animated style suits text-heavy posts where the video is background mood, not the focus. Realistic (LivePortrait) suits posts where Cordelia IS the focus.
