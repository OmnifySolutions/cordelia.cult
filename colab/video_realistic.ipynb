---
## Notes

- **`--still` flag (SadTalker):** keeps body motion minimal so Cordelia doesn't sway unnaturally. Good for stationary talking-to-camera posts.
- **`--enhancer gfpgan` (SadTalker):** applies GFPGAN face restoration after lip sync. Makes the output sharper. Remove if you get errors.
- **Driving video options (LivePortrait):** the repo ships with `d0.mp4` through `d5.mp4`. `d0` = subtle, natural. `d3`/`d4` = more expressive. Try different ones to see what suits the post.
